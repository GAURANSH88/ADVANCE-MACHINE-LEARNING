```python
def check_missing_value(df):
    missing_features=[]
    for i in df.columns:
        temp = df[i].isnull().sum()
        if temp!=0:
            print(i," feature has ",temp," missing value")
            missing_features.append(i)
    return missing_features

def check_features_datatype(df):
    categorical_feature = [i  for i in df.columns if df[i].dtype == 'O']
    numerical_feature = [i  for i in df.columns if df[i].dtype!= 'O']
    return categorical_feature,numerical_feature

def _check_datatype_is_string(df,feature):
    if df[feature] == 'O':
        return True
    else:
        return False
    
```

```python
categorical_features , numerical_features = check_features_datatype(df)
```

```python
print("Features where year is involved : ")
year_feature =[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature ]
year_feature
```

```python
df.drop(['Id'],axis=1,inplace=True)
```

```python
for i in missing_features:
    if i in numerical_features:
        temp=df[i].mean()
        df[i].fillna(temp,inplace=True)
    if i in categorical_features:
        temp=df[i].mode()[0]
        df[i].fillna(temp,inplace=True)
```

```python
for i in categorical_features:
    dict={}
    count=1
    for j in df[i].unique():
        dict[j]=count
        count+=1
    df[i]=df[i].map(dict)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['SalePrice'],axis=1)
y=df['SalePrice']
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.feature_selection import mutual_info_regression , SelectPercentile
```

```python
mutual_info = mutual_info_regression(x_train , y_train)
mutual_info
```

```python
mutual_info = pd.Series(mutual_info)
mutual_info.index = x_train.columns
mutual_info.sort_values(ascending=False)
```

```python
mutual_info.head()
```

```python
mutual_info.sort_values(ascending=False).plot.bar(figsize=(15,5))
```

```python
from sklearn.feature_selection import SelectPercentile
```

```python
selected_top_columns = SelectPercentile(mutual_info_regression, percentile=60)
selected_top_columns.fit(x_train, y_train)
```

```python
selected_top_columns.get_support()
```

```python
selected_columns = x_train.columns[selected_top_columns.get_support()]
```

```python
selected_columns = x_train.columns[selected_top_columns.get_support()]
```

```python
selected_train_features = x_train[selected_columns]
selected_train_features
```

```python
selected_test_features = x_test[selected_columns]
selected_test_features
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score
```

```python
lasso = Lasso()
lasso.fit(selected_train_features , y_train)
pred = lasso.predict(selected_test_features)
print(r2_score(pred , y_test))
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge()
ridge.fit(selected_train_features, y_train)
pred = ridge.predict(selected_test_features)
print(r2_score(pred , y_test))
```

# Comparison of ridge and lasso
both ridge regression and Lasso find their respective advantages. Ridge regression does not completely eliminate (bring to zero) the coefficients in the model whereas lasso does this along with automatic variable selection for the model. This is where it gains the upper hand.

Both these techniques tackle overfitting, which is generally present in a realistic statistical model. It all depends on the computing power and data available to perform these techniques on a statistical software. Ridge regression is faster compared to lasso but then again lasso has the advantage of completely reducing unnecessary parameters in the model.

there is not much difference in accuracy of lasso and ridge , lasso has accuracy of 0.7650 and ridge has 0.7643

lasso has ability to do feature selection , features which are of no use and are not contributing much to predict target value are removed

features which are less contributing to predict target value will eventually effect the model , Often neither one is overall better. Lasso can set some coefficients to zero, thus performing variable selection, while ridge regression cannot. sometimes performing variable selection will help model to increase in accuracy and sometimes it can result in slight reduction of accuracy because even those features are contributing somewhat to predict target variable

here in this case there is slight increase in accuracy of lasso compared to ridge
